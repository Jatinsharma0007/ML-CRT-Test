# Answer 17,18,19

```python
data = pd.read_csv("Book1.csv")
```

```python
data.head()
```

# answer 17

```python
# the average salary using numpy
avg_salary = np.mean(data["Salary"])
print("Average salary:", avg_salary)
```

# answer 18

```python
# Filtering out employees with salary greater than 60000
filtered_data = data[data["Salary"] > 60000]
print(filtered_data)
```

# answer 19

```python
# Ploting a bar chart representing the department-wise average salary using Matplotlib

# Calculate the average salary for each department
avg_salary_by_department = data.groupby("Department")["Salary"].mean()

# Create a bar chart
plt.bar(avg_salary_by_department.index, avg_salary_by_department.values)
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Average Salary by Department")
plt.show()
```

# Answers 20,21,22

```python
emp = pd.read_csv("employeedata.csv")
```

# Answer 20

```python
#Ploting a bar chart to visualize the number of employees in each department using Matplotlib

# Count the number of employees in each department
emp_count_by_department = emp["Department"].value_counts()

# Create a bar chart
plt.bar(emp_count_by_department.index, emp_count_by_department.values)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees by Department")
plt.show()
```

# Answer 21

```python
# Filter out employees with a salary greater than 60,000 using Pandas
filtered_data = emp[emp["Salary"] > 60000]
print(filtered_data)
```

# Answer 22

```python
# Create a line graph for the salary trend by department
plt.figure(figsize=(10,6))
plt.plot(emp.groupby("Department")["Salary"].mean())
plt.xlabel("Department")
plt.ylabel("Average Salary")
```
